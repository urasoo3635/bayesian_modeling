```python
import pandas as pd
import numpy as np
import os
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
```

### データの読み込みと前処理（標準化）

```python
# カルフォルニア住宅価格のデータセットをロード
datapath = "./california_housing.csv"
if os.path.exists(datapath):
    df = pd.read_csv(datapath)
    df = df.drop(columns="Unnamed: 0")
else:
    calif = fetch_california_housing()
    df = pd.DataFrame(calif.data, columns = calif.feature_names).assign(Price=np.array(calif.target))

df.describe()
```
